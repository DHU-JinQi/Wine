```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.optimizers import SGD
from keras.utils import np_utils

import numpy as np
import pandas as pd
import time

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv(r'Wine.csv')
```

```python
scaler=MinMaxScaler()
scaler.fit(data)
scaled_data=scaler.transform(data)
X=scaled_data[:,0:13]
Y=data['quality'].as_matrix()
X_values = X.astype('float32')
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_values,Y,test_size=0.3,random_state=23)
```

```python
encoder_y_train = LabelEncoder()
encoded_y_train = encoder_y_train.fit_transform(y_train)
dummy_y_train = np_utils.to_categorical(encoded_y_train)

encoder_y_test = LabelEncoder()
encoded_y_test = encoder_y_test.fit_transform(y_test)
dummy_y_test = np_utils.to_categorical(encoded_y_test)
```

```python
model1_1 = Sequential()
model1_1.add(Dense(64, input_shape=(13,), activation='relu'))
model1_1.add(Dropout(0.2))
model1_1.add(Dense(32, activation='relu'))
model1_1.add(Dense(16, activation='relu'))
model1_1.add(Dense(7, activation='softmax'))
model1_1.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model1_1.summary()
```

```python
np.random.seed(10000)
train1_1 = model1_1.fit(x_train, dummy_y_train, batch_size=1024, epochs=2000, verbose=1 ,validation_split=0.2,shuffle=True)
```

```python
pyplot.plot(train1_1.history['loss'], label='training loss')
pyplot.plot(train1_1.history['val_loss'], label='validation loss')
pyplot.legend()
pyplot.show()
```

```python
min_loss_index=train1_1.history['val_loss'].index(min(train1_1.history['val_loss']))
train1_1.history['val_acc'][min_loss_index]
```
